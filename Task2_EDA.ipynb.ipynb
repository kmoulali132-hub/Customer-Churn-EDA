```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Settings
%matplotlib inline
sns.set_style("whitegrid")

# Load Dataset
df = pd.read_csv("Data/Cleaned_Telco_Customer_Churn.csv")

print("="*60)
print("DATASET LOADED SUCCESSFULLY")
print("="*60)

print("\nFirst 5 Rows:")
display(df.head())

print("\nDataset Shape:", df.shape)

print("\nDataset Information:")
df.info()

print("\nMissing Values:")
print(df.isnull().sum())

print("\nSummary Statistics:")
display(df.describe())

print("\nColumns:")
print(df.columns.tolist())

# Overall Churn Rate
churn_rate = df["Churn_Yes"].mean() * 100
print(f"\nOverall Customer Churn Rate: {churn_rate:.2f}%")

# ------------------- Visualizations -------------------

# Churn Distribution
plt.figure(figsize=(6,4))
sns.countplot(x=df["Churn_Yes"])
plt.xticks([0,1],["No","Yes"])
plt.title("Customer Churn Distribution")
plt.show()

# Gender Distribution
plt.figure(figsize=(6,4))
sns.countplot(x=df["gender_Male"])
plt.xticks([0,1],["Female","Male"])
plt.title("Gender Distribution")
plt.show()

# Senior Citizen
plt.figure(figsize=(6,4))
sns.countplot(x=df["SeniorCitizen"])
plt.xticks([0,1],["No","Yes"])
plt.title("Senior Citizen Distribution")
plt.show()

# Partner
plt.figure(figsize=(6,4))
sns.countplot(x=df["Partner_Yes"])
plt.xticks([0,1],["No","Yes"])
plt.title("Partner Distribution")
plt.show()

# Dependents
plt.figure(figsize=(6,4))
sns.countplot(x=df["Dependents_Yes"])
plt.xticks([0,1],["No","Yes"])
plt.title("Dependents Distribution")
plt.show()

# Tenure
plt.figure(figsize=(8,4))
sns.histplot(df["tenure"], bins=30, kde=True)
plt.title("Customer Tenure Distribution")
plt.show()

# Monthly Charges
plt.figure(figsize=(8,4))
sns.histplot(df["MonthlyCharges"], bins=30, kde=True)
plt.title("Monthly Charges Distribution")
plt.show()

# Total Charges
plt.figure(figsize=(8,4))
sns.histplot(df["TotalCharges"], bins=30, kde=True)
plt.title("Total Charges Distribution")
plt.show()

# Monthly Charges vs Churn
plt.figure(figsize=(8,4))
sns.boxplot(x=df["Churn_Yes"], y=df["MonthlyCharges"])
plt.xticks([0,1],["No","Yes"])
plt.title("Monthly Charges vs Churn")
plt.show()

# Total Charges vs Churn
plt.figure(figsize=(8,4))
sns.boxplot(x=df["Churn_Yes"], y=df["TotalCharges"])
plt.xticks([0,1],["No","Yes"])
plt.title("Total Charges vs Churn")
plt.show()

# Contract Type vs Churn
contract = pd.DataFrame({
    "Month-to-Month": (~(df["Contract_One year"] | df["Contract_Two year"])).astype(int),
    "One Year": df["Contract_One year"].astype(int),
    "Two Year": df["Contract_Two year"].astype(int),
    "Churn": df["Churn_Yes"].astype(int)
})

contract.groupby("Churn")[["Month-to-Month","One Year","Two Year"]].sum().T.plot(
    kind="bar", figsize=(8,5)
)
plt.title("Contract Type vs Churn")
plt.xlabel("Contract Type")
plt.ylabel("Customers")
plt.legend(["No Churn","Churn"])
plt.show()

# Payment Method vs Churn
payment = pd.DataFrame({
    "Electronic Check": df["PaymentMethod_Electronic check"].astype(int),
    "Credit Card": df["PaymentMethod_Credit card (automatic)"].astype(int),
    "Mailed Check": df["PaymentMethod_Mailed check"].astype(int),
    "Other": (~(
        df["PaymentMethod_Electronic check"] |
        df["PaymentMethod_Credit card (automatic)"] |
        df["PaymentMethod_Mailed check"]
    )).astype(int),
    "Churn": df["Churn_Yes"].astype(int)
})

payment.groupby("Churn")[["Electronic Check","Credit Card","Mailed Check","Other"]].sum().T.plot(
    kind="bar", figsize=(8,5)
)
plt.title("Payment Method vs Churn")
plt.xlabel("Payment Method")
plt.ylabel("Customers")
plt.legend(["No Churn","Churn"])
plt.show()

# Correlation Heatmap
plt.figure(figsize=(18,14))
sns.heatmap(df.corr(), cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()

# Top Correlations with Churn
print("\nTop Correlations with Churn:")
print(df.corr()["Churn_Yes"].sort_values(ascending=False).head(10))

print("\n" + "="*60)
print("EDA COMPLETED SUCCESSFULLY")
print("="*60)

print("""
Key Observations:
1. Overall churn rate is approximately {:.2f}%.
2. Most customers are not senior citizens.
3. Customers with month-to-month contracts have higher churn.
4. Electronic check users tend to churn more.
5. Longer-tenure customers are less likely to churn.
6. Higher monthly charges are associated with higher churn.
7. Two-year contract customers show the lowest churn.
""".format(churn_rate))
```
